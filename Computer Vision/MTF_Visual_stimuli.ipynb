```python
from google.colab import files
import matplotlib.pyplot as plt
import math
from PIL import Image
 
a,b,c = 3.0, 1, 0.015 #constants

#Function that returns an image object with w and h parameters
def create_image(width , height):
  return Image.new("L",size = (width,height), color = 0)

def merge_exp_function(w):#w(x) exp function to get Frequency
  return a * math.exp(b*w)

def vertical_axis(y): #Y exp axis function
  return math.exp(-c*y)
  
def create_visual_stimuli():
  new_image = create_image(500,500)
  for i in range(new_image.width):
    aux = function_f(i/100.0)
    for j in range(new_image.height):
      if i != 0:
        new_image.putpixel((i,j), 100 + int(vertical_axis(new_image.height-j)*(100*aux)))
  plt.imshow(new_image)
  new_image.save('stimuli.jpg')
  files.download('stimuli.jpg')

def function_f(x): #f(x) = sin(ae^bx) sinusoidal function
  return math.sin(merge_exp_function(x))
  
create_visual_stimuli()
```
